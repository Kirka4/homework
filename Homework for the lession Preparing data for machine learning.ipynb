```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
df =pd.read_csv(r"C:\Users\kir4i\Downloads\Telegram Desktop\3.4 dataset.csv")
```

```python
# Кодирование категориальных данных
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Pet'] = le.fit_transform(df['Pet'])
df['Country'] = le.fit_transform(df['Country'])
```

```python
df['Gender_Pet'] = df['Gender'] + df['Pet']
df['Gender_Age'] = df['Gender'] + df['Age']
df['Gender_Country'] = df['Gender'] + df['Country']
df['Pet_Age'] = df['Pet'] + df['Age']
df['Pet_Country'] = df['Pet'] + df['Country']
df['Age_Country'] = df['Age'] + df['Country']
```

```python
df['calculated_column'] = df['Gender'] * df['Age'] * df['Pet']* df['Country']
```

```python
kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(df.drop(['Name'], axis=1))
```
